** Splitting the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
ad_data.head()
```

```python
X = ad_data.drop(['Ad Topic Line','City','Country','Timestamp','Clicked on Ad'],axis=1)
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
model_1 =LogisticRegression(max_iter=500)
```

```python
model_1.fit(X_train,y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
model1_pred = model_1.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, model1_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, model1_pred))
```
